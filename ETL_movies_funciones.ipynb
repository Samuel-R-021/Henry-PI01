# Dataset funciones 1-4

```python
funciones_1_4_dataset = movies[['title','popularity','vote_average','vote_count']].copy()
funciones_1_4_dataset
```

```python
funciones_1_4_dataset['release_year'] = movies['release_date'].dt.year
funciones_1_4_dataset['month_name'] = movies['release_date'].dt.month_name(locale='es_VE.utf8')
funciones_1_4_dataset['day_name'] = movies['release_date'].dt.day_name(locale='es_VE.utf8')

funciones_1_4_dataset
```

```python
funciones_1_4_dataset['lowercase_title']=funciones_1_4_dataset['title'].str.lower()
# se cambia los dias con acento que generan error por los correctos
funciones_1_4_dataset['day_name'].replace('Sã¡bado', 'Sábado', inplace=True)
funciones_1_4_dataset['day_name'].replace('Miã©rcoles', 'Miércoles', inplace=True)
funciones_1_4_dataset
```
