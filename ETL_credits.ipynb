# cast_credits transformacion

```python
cast_credits = cast_credits.explode('cast',ignore_index= True)
cast_credits
```

```python
expanded_cast = pd.json_normalize(cast_credits['cast'])
expanded_cast
```

```python
# para la funcion 5 get_actor(), luego de expandir, solo nos quedamos con el id, name y order
cast_credits = pd.concat([cast_credits['id'],expanded_cast[['name','order']]], axis=1)
cast_credits.head(3)
```

```python
cast_credits.shape
```

```python
# estamos considerando solo aquellas peliculas donde el actor esta entre los primeros 5 en el order de protagonismo
cast_credits=cast_credits[cast_credits['order'].isin([0,1,2,3,4])]
cast_credits.shape
```

```python
cast_credits.sort_values(by='name')
```

```python
# Creando columna lowercase_name para el input y la funcion
# nos ayudara a buscar mas facilmente

cast_credits['lowercase_name']=cast_credits['name'].str.lower()
cast_credits
```

# crew_credits transformacion

```python
crew_credits = crew_credits.explode('crew',ignore_index= True)
crew_credits
```

```python
expanded_crew = pd.json_normalize(crew_credits['crew'])
expanded_crew
```

```python
crew_credits = pd.concat([crew_credits['id'],expanded_crew[['job','name']]], axis=1)
crew_credits.head(3)
```

```python
# solo queremos los nombres de lso directores
crew_credits=crew_credits[crew_credits['job']=='Director']
crew_credits
```

```python
# similar a get_actor(), en get_director() usar el nombre como indice es mas facil
crew_credits['lowercase_name']=crew_credits['name'].str.lower()
crew_credits
```

```python
# eliminamos job que ya no lo necesitamos
crew_credits.drop(columns='job',inplace=True)
crew_credits
```
